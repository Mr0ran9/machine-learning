```python
x = alibi[['diameter', 'weight', 'red', 'green', 'blue']]
y = alibi[['name']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1 ,random_state=0)
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform (x_test)
```

```python
model7 = RandomForestClassifier(max_depth=5)
model7.fit(x_train, y_train)
```

```python
train_score = model7.score(x_train, y_train)
testing_score = model7.score(x_test, y_test)
print(f"Skor Training = {train_score:.4f}")
print(f"Skor Testing = {testing_score:.4f}")
```
